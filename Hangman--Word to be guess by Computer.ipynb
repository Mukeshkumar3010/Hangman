```python
# Hangman game
#

# -----------------------------------
# Helper code
'''
creating unique dictionay with key as int and the value will be unique letters from the list(lst)
guesslst is used not to considered the word which is already guessed
'''
def new_UniqueLetter_Dic(lst,guesslst):
    lst = set("".join(lst))-set(guesslst)
    return dict(enumerate(lst))

'''
Below is to update the WordList based on guess letter here cnf will check user input whether they have
Entered 1(if the guess is present in there word) or 0 (if the guess is not present in the word)
using reverse in order to execute faster, frequency is being used to check how many times guess is 
present in the word, None is default argument when we are calling this function first time.
pos is used to update the list based on the position of letter
'''

def Update_WordsList(cnf,guess,lst1,frequency=None,pos=None):
    if cnf:
        for eachlist in reversed(lst1):
            if guess not in eachlist:
                lst1.remove(eachlist)
            
            try:
                if eachlist.count(guess) < frequency:  # count will count no of guess(letter) present in index
                    lst1.remove(eachlist)              # i.e good(index)-- o(guess)-- count of o is 2. remove that word less than 2

                else:
                    if not eachlist[pos] == guess:
                        lst1.remove(eachlist)

            except:
                pass
    else:
        for eachlist in reversed(lst1):
            if guess in eachlist:
                lst1.remove(eachlist)
    return lst1
'''
To check if we have only one element then ask user to confirm 
if the word is same as he/she thought and return True/False accordingly
'''
def lencheck(OnlyOneElm):
    
    if len(OnlyOneElm) == 1:
        
        print("Is your word : " + ", ".join(wordsList[length]))
        userin = int(input("Press \"1\" for yes, or \"0\" for no: "))
        
        if userin == 1:
            print("I win! Try harder next time!")
            return True
        else:
            wordsList[length] = []
            return False
    return False


# end of helper code
# -----------------------------------

# Main Code

import random
#Open the txt file as read only( should be in same path where this points to else give full path)
print("Loading word list from file...")
words = open("Words_C.txt", "r")

#now we will take words variable and read one by one and put them into single list
originalList = words.read().splitlines()

print("  ", len(originalList), "words loaded.")
wordsList={}

'''
creat a dictionary with key as length of word and club words with same length as value for key as list
this will help in accessing the list for lenght which user thinks of word
'''
for elm in originalList:
    wordsList.setdefault(len(elm),[]).append(elm)
    
numOfGuesses = 10
length = int(input("Think of a word.\nHow many letters are in your word?: "))
word = "." * length
confirmWord = list(word)
All_guess = []

'''
Creates number system for user for each dot in word as below
 ....
 1234
 
'''
numberSystem = ""
numberSyst = list(range(1, length + 1))
for number in numberSyst:
    numberSystem += str(number)
    
#Game will start
_run = True
while _run:
    #Decides word is not in list. Adds word to text file
    if len(wordsList[length]) == 0:
        print("I give up! You win!")
        userWord = input("Please enter your word in lower-case letters to add in Database: ")
        if userWord not in originalList:
            addition = open("words.txt", 'a')
            addition.write(userWord + "\n")
            addition.close()
        print("Thank you for contributing!")
        break
        
    letterCheck = ""
    print("I have " + str(numOfGuesses) + " guesses left.")
    print("----------------------------")
    #Finds letters not in wordsList[length] and delete letters
    letters = new_UniqueLetter_Dic(wordsList[length],All_guess)
    #Letter guesser
    guessInt = random.randint(0, len(letters)-1)
    guess = letters[guessInt]
    All_guess.append(guess)
    
    #Game time
    print("Is there a(n) " + guess + " in the word?")
    confirmation = int(input("Press \"1\" for yes, or \"0\" for no: "))
    
    if confirmation == 1:
        #Calling function to check if we have more than 1 element in list
        if lencheck(wordsList[length]):
            break
  
        while confirmation == 1:
            print(word)
            print(numberSystem)
            position = int(input("Choose the number that represents the position of the letter: ")) - 1
            
            confirmWord[position]=guess
            word = "".join(confirmWord)
            freq = word.count(guess)
            
            #Delete words which does not matches no of specific letter and postion of letter
            Update_WordsList(confirmation,guess,wordsList[length],freq,position)

            if lencheck(wordsList[length]):
                _run = False
                break
            
            print(word)
            print(numberSystem)
            
            if "." not in confirmWord:
                print("Your word is : " + ", ".join(wordsList[length]) + "\nI win! Try harder next time!")
                _run = False
                break
                
            print("Is there another " + guess + " in the word?")
            confirmation = int(input("Press \"1\" for yes, or \"0\" for no: "))
            
    elif confirmation == 0 and numOfGuesses != 1:
        #Delete words with wrong guess
        Update_WordsList(confirmation,guess,wordsList[length])
        
        if lencheck(wordsList[length]):
            _run = False
            break

        numOfGuesses -= 1
    else:
        #Nobody's perfect
        print("I lose! You're just too good!")
        break
```
